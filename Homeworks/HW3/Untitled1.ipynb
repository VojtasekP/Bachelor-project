```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
data_count = 1000

DEVICE = 'cuda'
freq = 10*torch.rand(data_count, 1, device=DEVICE) + 1
random_shift = torch.normal(0, 10, size=(data_count, 1), device=DEVICE)
points = torch.arange(0,6,1/50, device=DEVICE).repeat(data_count, 1) + random_shift
print(points)
random_phase_modulator = torch.normal(0, 5, size=(data_count, 1), device=DEVICE)
data_matrix = torch.sin(points * freq + random_phase_modulator)
for i in range(len(data_matrix[0])):
    random_noise = torch.normal(0, 0.04, size=(1, len(data_matrix[0])), device=DEVICE)
    data_matrix[i] = data_matrix[i] + random_noise
```
